# Remove on base of old and new list

## keep old and add new at end

```python
# importing pandas library
import pandas as pd

# creating and initializing a nested list
students = [['jackma', 34, 'Sydeny', 'Australia','old'],
			['Ritika', 30, 'Delhi', 'India','old'],
			['Vansh', 31, 'Delhi', 'India','old'],
			['Nany', 32, 'Tokyo', 'Japan','old']]

# Create a DataFrame object
old_df = pd.DataFrame(students,
				columns=['Name', 'Age', 'City', 'Country','List']) #,

old_df
```

```python
# importing pandas library
import pandas as pd

# creating and initializing a nested list
students1 = [['jackma', 34, 'Sydeny', 'Australia','New'],
			['Ritika', 30, 'Delhi', 'India','New'],
			['Vansh', 31, 'aa', 'India','New'],
			['Nany', 32, 'aa', 'Japan','New'],
			['May', 16, 'New York', 'US','New'],
			['Michael', 17, 'las vegas', 'US','New']]

# Create a DataFrame object
new_df = pd.DataFrame(students1,
				columns=['Name', 'Age', 'City', 'Country','List']) #,

new_df
```

```python
combine_df=pd.concat([old_df,new_df])
combine_df
```

```python
unique_df= combine_df.drop_duplicates(['Name'])
unique_df
```

## keep new and remove old

```python
keep_new_df=combine_df.drop_duplicates(['Name'],keep="last",)
```
